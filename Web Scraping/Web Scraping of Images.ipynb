The function that is invoked. It uses beautiful Soup library

```python
import urllib.request as url
from bs4 import BeautifulSoup
import requests
import shutil
import os, sys

def downloadAllImages(celeb,fromPage,numberOfPages):
    #the url of the 
    urli = 'http://celebmafia.com/' +celeb+'/'
    os.mkdir(celeb)
    os.chdir(celeb)
    for i in range(fromPage,fromPage+numberOfPages+1):
        os.mkdir(str(i))
        os.chdir(str(i))
        wiki = urli+"page/"+str(i)+'/'
        #this is the format of url for each of the pages
        r = url.Request(wiki, headers= {'User-Agent' : 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36'})
        html = url.urlopen(r)
        soup = BeautifulSoup(html)
        
        #this list will be used to store all the links with "more-link", this is where we drill down to other pictures
        allPageLinks =[]
        for nextlink1 in soup.findAll("a", {"class" : "more-link"}):
            a = nextlink1.get('href')
            allPageLinks.append(a)
        #print(os.getcwd())
        
        #for all the image links on that page. download the image
        for pagelink in allPageLinks:
            imgURL = pagelink
            r1 = url.Request(imgURL, headers= {'User-Agent' : 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36'})
            html1 = url.urlopen(r1)
            soup = BeautifulSoup(html1)
            allimageLinks =[]
            for nextlink1 in soup.findAll("a"):
                a = nextlink1.get('href')
                allimageLinks.append(a)
            lo = [s for s in allimageLinks if '.jpg' in s]
            
            
            #once all the image link has been fetched , write the image files to the local disk
            for image in lo:
                r = requests.get(image, stream=True)
                if r.status_code == 200:
                    print(celeb + str(allPageLinks.index(pagelink))+ str(lo.index(image)) +'.jpg')
                    with open(celeb +  str(allPageLinks.index(pagelink))+ str(lo.index(image)) +'.jpg', 'wb') as f:
                        r.raw.decode_content = True
                        shutil.copyfileobj(r.raw, f)
                        f.close()
                      
        
        os.chdir("..")
```
