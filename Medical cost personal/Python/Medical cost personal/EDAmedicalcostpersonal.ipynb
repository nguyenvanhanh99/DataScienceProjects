```python
# Importing the libraries
import numpy as np
import matplotlib
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt

# Importing the dataset
dataset = pd.read_csv('insurance.csv')
```

```python
#Check the value counts of all the categorical data types
for value in dataset.columns:
    if dataset[value].dtype == object:
        print(dataset[value].value_counts())
```

```python
%matplotlib inline
```

```python
#takes input in the form of list(category) and returns output 
#result showing each of the other sub category values that makes 
#up the given category

def valueCountsForSubcategory(category,quant):
    for subs in category:
        valtype = dataset[subs].value_counts().index #get all the values in each category
        
        for val in valtype:
            print('#####:'+ val + ':####' )
            plotData = dataset[dataset[subs]==val]   #for each category value filter the dataset
            for value in plotData.columns:
                if plotData[value].dtype == object:  # check if the datatype is of object
                    print(plotData[value].value_counts())
                    for quantcat in quant:
                        sns.catplot(x=value, y=quantcat, kind="box", data=plotData);
```

```python
def PlotCorr(X):
    corr = X.corr()
    cmap = sns.diverging_palette( 150 , 275 , as_cmap = True )
    sns.heatmap(
        corr, cmap = cmap, square = True, cbar = True,
     annot = True, annot_kws = { 'fontsize' : 12}
    )
    plt.yticks(rotation = 0)
    plt.xticks(rotation = 0)
    plt.title("Correlation Matrix")
```

```python
PlotCorr(dataset)
```
