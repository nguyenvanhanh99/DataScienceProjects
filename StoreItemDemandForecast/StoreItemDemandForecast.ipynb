```python
storeDataset=[]
for i in range(1,12):
    print(i)
    storeDataset.append(dataset[dataset.store==(i-1)])
```

```python
jointSaleDataset = pd.DataFrame(storeDataset[1])
columnNames =[]
for i in range(1,11):
    columnNames.append('store'+str(i))
    columnNames.append('item'+str(i))
    columnNames.append('sales'+str(i))
print(columnNames)

for j in range(1,10):
    print(j)
    jointSaleDataset = pd.concat([jointSaleDataset,storeDataset[j+1]], axis=1)
jointSaleDataset.columns = columnNames
```
