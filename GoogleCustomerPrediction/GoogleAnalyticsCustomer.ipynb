Identify the keys that comes as part of the dictionary set in some of the attributes like device , totals, geoNetwork and traffic source

```python
import json
d_device =json.loads(dataset['device'][1])
d_totals =json.loads(dataset['totals'][1])
d_geoNetwork =json.loads(dataset['geoNetwork'][1])
d_trafficSource =json.loads(dataset['trafficSource'][1])
```

```python
print(d_device.keys())
print(d_totals.keys())
print(d_geoNetwork.keys())
print(d_trafficSource.keys())

attributesInd_device =list(d_device.keys())
attributesInd_totals =list(d_totals.keys())
attributesInd_geoNetwork =list(d_geoNetwork.keys())
attributesInd_trafficSource =list(d_trafficSource.keys())
```

```python
print(dataset['geoNetwork'][1])
print(dataset['trafficSource'][1])
```

Create a subset of data to check the piece of code that can be used to break the device field into multiple columns 

```python
df = dataset.iloc[1:1000,]
```

```python
df.describe()
```

```python
len(attributesInd_device) + len(attributesInd_totals) + len(attributesInd_geoNetwork) + len(attributesInd_trafficSource)+dataset.shape[1]
```

```python
for attr in attributesInd_device:
    df[attr] = df['device'].apply(lambda x:(json.loads(x)).get(attr))
```

```python
for attr in attributesInd_totals:
    df[attr] = df['totals'].apply(lambda x:(json.loads(x)).get(attr))
```

```python
for attr in attributesInd_geoNetwork:
    df[attr] = df['geoNetwork'].apply(lambda x:(json.loads(x)).get(attr))
```

```python
for attr in attributesInd_trafficSource:
    df[attr] = df['trafficSource'].apply(lambda x:(json.loads(x)).get(attr))
```

```python
df = df.drop('trafficSource', 1)
df = df.drop('geoNetwork', 1)
df = df.drop('totals', 1)
df = df.drop('device', 1)
```

Using the same processing on the main dataset

```python
for attr in attributesInd_device:
    print('device:'+ attr)
    dataset[attr] = dataset['device'].apply(lambda x:(json.loads(x)).get(attr))
for attr in attributesInd_totals:
    print('totals:'+ attr)
    dataset[attr] = dataset['totals'].apply(lambda x:(json.loads(x)).get(attr))
for attr in attributesInd_geoNetwork:
    print('geoNetwork:'+ attr)
    dataset[attr] = dataset['geoNetwork'].apply(lambda x:(json.loads(x)).get(attr))
for attr in attributesInd_trafficSource:
    print('trafficSource:'+ attr)
    dataset[attr] = dataset['trafficSource'].apply(lambda x:(json.loads(x)).get(attr))
```
